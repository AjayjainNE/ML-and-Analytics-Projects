Given the following NumPy array: <br> <br>

##### Write a Python program to do the following tasks: <br>
```` a) Print the last 3 elements of the array using the slicing technique.```` <br>

`````python
import numpy as np
numbers = np.array([200, 150, 350, 40, 570, 50, 220, 990, 40, 444])

print (numbers[-3:])    #Here we used '-' sign as we are asked to print last 3 elements ':' for slicing and '3' for the index mapping. 
`````

```` b) Print every other value in the array.```` <br>

`````python
import numpy as np
numbers = np.array([200, 150, 350, 40, 570, 50, 220, 990, 40, 444])

print (numbers [::2])   #Here we used '::' to count the every other array starting from first position and '2' to count every second step.
`````

```` c) Find the largest number in the array, and its index position. Print the results.```` 

`````python
largest = np.max(numbers)     #This will calculate the largest value.
position = np.argmax (numbers)     #This will let us know the index of value.

print("largest number in the array is : ", largest)
print("the index of the number is : ", position)
`````

```` d) Find the smallest number in the array. If there are duplicates, print all the indices of the smallest number. ````

`````python
smallest = np.min(numbers)     #To find the smallest number in given array.
indices = np.where (numbers == smallest)[0]     #To know the indices where this value occured.

print("The smallest number in the array :",smallest)
print("The indices of smallest number", indices)
`````

```` e) Compute and print the average of all numbers in the array.````

`````python
average = np.mean(numbers)     #this will output the mean of all numbers in array.

print("Average of all numbers in array is :", average)
`````

```` f) Print all numbers that are greater than 250.````

`````python
print(numbers[numbers > 250])    #this will search array and output all numbers greater than 250.
`````

```` g) Remove the value 40 from the array. Print the new array. Use this new array for the rest of the exercise.````

`````python
new_numbers = numbers[numbers !=40]    #This will remove value 40 from the array and create new array as 'new_numbers'

print ("The new array is", new_numbers)
`````

```` h) Add the number 1000 to the end of the array. Use this new array for the rest of the exercise.````

`````python
import numpy as np

new_numbers = np.array ([200, 150, 350, 570,  50, 220, 990, 444])    #Updated array after removing all '40'

update_numbers = np.append (new_numbers, 1000)    #append function will add '1000' to the list of array by default in the end

print("The new array after adding 1000 is :", update_numbers)
`````

```` i) Replace all numbers less than 100 with 10. Use this new array for the rest of the exercise.````

`````python
import numpy as np

update_numbers = np.array ([200,  150,  350,  570,   50,  220,  990,  444, 1000])    #new array we get after adding 1000 in the end

update_numbers [update_numbers < 100 ] = 10    #this will update the numbers '< 100' with the new value '10'

print ("New list of array is :", update_numbers)
`````

```` j) Increase each number in the array by 10%. Print the new array. ````

`````python
import numpy as np
update_numbers = np.array ([ 200,  150,  350,  570,   10,  220,  990,  444, 1000])    #New list of array after replacing all numbers < 100 to 10

increased_numbers = update_numbers * 1.1    #this will increase each number in array by 10% 

print("New list of array after increase of 10% :", increased_numbers)
`````
